```python
##Exporting Data with SL API
# Author: Santosh Kumar Udayakumar
# Script :Inspect and explore the SL API and SL Netex file and assess what is the best data source and download the data files to the source directory 

import requests
import zipfile
import io
import pandas as pd

# Replace with your actual API key
api_key = '<REPLACE_API_KEY>'
url = f'https://opendata.samtrafiken.se/gtfs-sweden/sweden.zip?key={api_key}'
response = requests.get(url)
if response.status_code == 200:
    # Step 2: Extract the ZIP file
    with zipfile.ZipFile(io.BytesIO(response.content)) as z:
        z.extractall('gtfs_data')  # Extract to a folder named 'gtfs_data'
else:
    print("Failed to download the data:", response.status_code)
```

```python
###Lookup latest SL data and Place in the source Directory

import requests

# Replace with your actual operator and API key
operator = 'sl'  # Example operator
apikey = '<Replace API KEY>'        # Example API key

# Construct the URL
url = f"https://opendata.samtrafiken.se/netex/{operator}/{operator}.zip?key={apikey}"

# Make the GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Save the content to a file
    with open(f"{operator}.zip", 'wb') as file:
        file.write(response.content)
    print(f"Downloaded {operator}.zip")
else:
    print(f"Failed to retrieve data: {response.status_code} - {response.text}")
```
